#2. Data Cleaning

```python
#Drop duplicates, fill/missing values:
df.drop_duplicates(inplace=True)
df.isna().sum()
```

# 3. Exploratory Data Analysis (20 Questions + Code)

```python
#Basic Info
df.describe()
```

## Sales & Profit

```python
#	What are the total sales and total profit for the Superstore?
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
print(f"Total Sales: ${total_sales:,.2f}")
print(f"Total Profit: ${total_profit:,.2f}")
```

```python
#	What is the average order value?
avg_order_value = df.groupby('Order ID')['Sales'].sum().mean()
print(f"Average Order Value: ${avg_order_value:,.2f}")
```

```python
# Which product has the highest sales? Which has the highest profit?
highest_sales_product = df.loc[df['Sales'].idxmax(), 'Product Name']
highest_profit_product = df.loc[df['Profit'].idxmax(), 'Product Name']
print(f"\nProduct with highest sales: {highest_sales_product}")
print(f"Product with highest profit: {highest_profit_product}")
```

```python
#	Which product category (Furniture, Technology, Office Supplies) is the most profitable?
# Most profitable category
category_profit = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
print("\nProfit by category:")
category_profit
```

## Category & Sub-Category

```python
# Which region has the highest sales? Which has the highest profit margin?
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

print("\nSales by region:")
region_sales

```

```python

# Calculate region profit margin
region_profit = df.groupby('Region')['Profit'].sum()
region_sales = df.groupby('Region')['Sales'].sum()
region_profit_margin = (region_profit / region_sales).sort_values(ascending=False)

print("Profit Margin by Region:")
region_profit_margin
```

```python
#Profit by sub-category →
df.groupby('Sub-Category')['Profit'].sum()
```

```python
#Revenue by region →
df.groupby('Region')[['Sales','Profit']].sum()
```

```python
#	Which city is the top contributor to sales?
city_sales = df.groupby('City')['Sales'].sum().sort_values(ascending=False)
print("\nTop 5 cities by sales:")
city_sales.head()
```

```python
# Average number of orders per customer
avg_orders = df.groupby('Customer ID')['Order ID'].nunique().mean()
print(f"\nAverage number of orders per customer: {avg_orders:.2f}")
```

```python

# Top 5 customers by spending
top_customers = df.groupby(['Customer ID', 'Customer Name'])['Sales'].sum().sort_values(ascending=False).head(5)
print("\nTop 5 customers by spending:")
top_customers
```

```python
# Average discount per product category
avg_discount = df.groupby('Category')['Discount'].mean().sort_values(ascending=False)
print("\nAverage discount by category:")
print(avg_discount)

```

```python
#Create a new column Profit Margin (Profit/Sales). Which sub-category has the lowest profit margin?
# Profit margin analysis
df['Profit Margin'] = df['Profit'] / df['Sales']

# Sub-category with lowest profit margin
subcat_profit_margin = df.groupby('Sub-Category')['Profit Margin'].mean().sort_values()
lowest_margin_subcat = subcat_profit_margin.idxmin()
print(f"\nSub-category with lowest profit margin: {lowest_margin_subcat}")
```
